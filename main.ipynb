```python
# Load dataset
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore', category=UserWarning)
df = pd.read_csv('data.csv')
df
```

```python
# convert database column names to lowercase

df.columns = df.columns.map(lambda x: x.lower())
```

```python
# Data types of each columns
df.dtypes
```

```python
# Drop unnecessary columns
df.drop(['timestamp', '12. what kind of business you want to start?', '14. how do your family members perceive your enterprential activities?'], axis=1, inplace=True)
df
```

```python
# prepare dataset, by replacing range values (e.g, 400-500), more than values (e.g., 30+), Below keywords (e.g., Below 50) for specific columns

# column names of the dataset that needs processing
labels = ['2. age', '4. amount of financial aid recieved ............taka', '6. number of family members', '9. present monthly income (in taka)']

# interate through rows of dataset
for index, row in df.iterrows():

    # iterate through filtering columns of dataset
    for label in labels:

        # more than condition check and replace value
        if '+' in row[label]:
            df.at[index, label] = row[label].replace('+', '')

        # range condition check and replace value
        if '-' in row[label]:
            val = row[label].split('-')
            df.at[index, label] = int(float(val[0])+float(val[1])/2)

        # below condition check and replace value
        if 'Below' in row[label]:
            df.at[index, label] = int(row[label].replace('Below ',''))
            # row ['2. age'] = sum(val) / 2

df
```

```python
# Encode string labels to numbers, to apply regression in future

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

# Columns to convert
labels = ['3. maritial status', '8. present status of the respondent', '10. what do you want to do with this financial aid?', '11. do you have any specific business model?', '13. as a women. what kind of barrier you expect to find while starting a business?', '15. do you think you have enough importance in your family?', '16. do you believe that after receieving this financial help your voice will be heard by your family members?', '17. do you have any authority making decisions?', '18. do you believe that you are financially independent at present ?', '19. does financial dependency curb your freedom of movement?', '21. do you believe that your social status will increase after receiveing the financial aid?', '22. are the local markets, farmer friendly?', '23. do you have any plan to reach your products to local markets?', '24. after 6 month, are they gain success?']

for label in labels:
    df[label] = labelencoder.fit_transform(df[label])
df
```

```python
# Split train and test dataset
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2)
```

```python
# columns to train the classifier with given input
f_train = df_train.drop('24. after 6 month, are they gain success?', axis=1)

# columns to test the classifier with given input
f_test = df_test.drop('24. after 6 month, are they gain success?', axis=1)
f_train
```

```python
# rows to test the classifiers output
s_train = df_train[['24. after 6 month, are they gain success?']].copy()

# rows to test if the classifier could give expected output
s_test = df_test[['24. after 6 month, are they gain success?']].copy()
s_test
```

```python
# function to test the classifiers accuracy percentage
def accuracy_test(classifier, test_name):
    correct = 0
    wrong = 0
    cols = f_test.columns.to_list()
    for item_index, item_row in df_test.iterrows():
        # make prediction for the test dataset row
        prediction = classifier.predict([item_row[cols]])

        # check if the prediction is the same as the actual value
        if prediction == item_row['24. after 6 month, are they gain success?']:  # 15 is a guess for what would be acceptable
            correct = correct + 1
        else:
            wrong = wrong + 1

    total = wrong+correct

    print(test_name+" Accuracy: ", end='')
    print("%.2f %%"%((correct/total)*100))
    print("Total: "+str(total)+"\tCorrect: "+str(correct)+"\tWrong: "+str(wrong))
```

```python
# Gaussian Algorithm

from sklearn.naive_bayes import GaussianNB

# Create classifier
g_clf = GaussianNB()

# Train classifier
g_clf.fit(f_train, np.ravel(s_train.values)).predict(f_train)

# Prediction the response
g_clf.predict(f_test)
```

```python
accuracy_test(g_clf, 'Gaussian Naive Bayes')
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

# Create classifier
rf_clf = RandomForestClassifier(n_jobs=2, random_state=0)

# Train classifier
rf_clf.fit(f_train, np.ravel(s_train.values))

# Prediction the response
rf_clf.predict(f_test)
```

```python
accuracy_test(rf_clf, 'Random Forest')
```

```python
# Suppport vector machine
from sklearn import svm

# Create classifier
svm_clf = svm.SVC(kernel='linear')

# Train classifier
svm_clf.fit(f_train, np.ravel(s_train.values))

# Predict the response
svm_clf.predict(f_test)
```

```python
accuracy_test(svm_clf, 'Support vector machine')
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier

# create classifier
tree_clf = DecisionTreeClassifier()

# train classifier
tree_clf.fit(f_train, np.ravel(s_train.values))

# predict response

tree_clf.predict(f_test)
```

```python
accuracy_test(tree_clf, 'Decision Tree Classifier')
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

#create classifier
knn_clf = KNeighborsClassifier(n_neighbors=7)

# train classifier
knn_clf.fit(f_train, np.ravel(s_train.values))

# prediction classifier
knn_clf.predict(f_test)
```

```python
accuracy_test(tree_clf, 'K-nearest neighbour algorithm')
```
